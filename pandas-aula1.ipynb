```python
countries = ['Brazil', 'China', 'UK', 'India', 'Canada', 'Russia','USA', 'Australia']
df_population = pd.read_csv('population.csv', header=None, index_col=0, names=countries)
df_population
```

```python
df_area = pd.read_csv('area.csv', header=None, names=countries)
df_area
```

```python
#Pais com maior area
sr_area = df_area.squeeze() #passa para series
sr_area.idxmax()
```

```python
sr_total_pop = df_population.sum(axis=1)
sr_total_pop
```

```python
#proporção da população em cada país ao longo dos anos
df_population.T*100/sr_total_pop  #reanspõem o dataframe para ficar do mesmo tamanho de colunas que existe na series
```

```python
#construa uma tabela de densidade populacional e exporte para um csv
dens_pop =df_population/sr_area
dens_pop
```
