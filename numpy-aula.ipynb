```python
import numpy as np
countries = np.array(['Brazil', 'China', 'UK', 'India', 'Canada', 'Russia', 'USA', 'Australia'])

population_by_year = np.genfromtxt('population.csv', delimiter=',', dtype='int64')
area = np.genfromtxt('area.csv', delimiter=',', dtype='int64')
```

```python
#o total da população de todos os países para cada ano
total_population = population_by_year[:,1:].sum(axis=1)
print('O total da população é:\n', total_population)
```

```python
#proporção da população em cada pais ao longo do ano
total_population = np.array([total_population]).T
population_by_year/total_population*100
```

```python
#construir a matriz de densidade populacional ( pessoas por km2)
t_area = area.reshape(1,8)
density_matrix = population_by_year[:,1:]/area #Uso todas as colunas, mas pego apenas as colunas que de fato tem os valores, deixando a de ano de lado
density_matrix 
```

```python
#mais denso de 1950
index_max = density_matrix[0].argmax()
countries[index_max]
```

```python
#mais denso em 2023
years_col = population_by_year[:, 0] #pego apenas os anos
index_2023 = np.where(years_col == 2023) # Realiza o filtro, selecionando o index de 2023

index_max = density_matrix[index_2023].argmax() #procura qual o index da maior densidade na linha de 2023
countries[index_max]
```

```python
#mais denso em 2100
years_col = population_by_year[:, 0]
index_2100 = np.where(years_col == 2100)

index_max = density_matrix[index_2100].argmax()
countries[index_max]
```
