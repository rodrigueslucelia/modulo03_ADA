```python
import pandas as pd

df = pd.read_csv("data_shark_tank_brazil.csv")
print(f'{df.head}\n')
print(f'{df.info()}\n')
print(f'{df.describe()}\n')
```

```python
#Em quantas empresas cada shark investiu  sozinho?
#Análise para verificar como está os nomes dos investidores
investidores = df['name_investors'].value_counts()
print(f'\n{investidores}\n')
#Como há alguns em que os shrks investiram em conjunto, serão desconsiderados
df_investidores = df[(df['name_investors'].str.count(r'\+') == 0) & (df['name_investors'] != 'Nenhum') & (df['name_investors'] != 'no investment')]
#Separar  os investidores 
df_investidores
```
